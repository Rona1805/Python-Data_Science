```python
# DataFrame kopyaladık.
borsaVeri2 = borsaVeri.copy()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
import xgboost as xgb
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
from sklearn import model_selection
```

```python
# Bağımlı ve bağımsız değişkenleri ayırdık.
X = borsaVeri2.drop(["borsadaki_piyasa_degeri"], axis = 1)
y = borsaVeri2["borsadaki_piyasa_degeri"]
```

```python
# Eğitim ve test verilerinin ayrılması. Testi % 50 tuttuk.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.50, random_state = 144)
```

```python
# GridSearchCv - Modelimizdeki en iyi parametreleri bulur.
parametreler = {"colsample_bytree":[0.4,0.5,0.6],
                "learning_rate":[0.01,0.02,0.09],
                "max_depth":[2,3,4,5,6],
                "n_estimators":[100,200,500,2000]}
```

```python
xgb = XGBRegressor()
```

```python
grid = GridSearchCV(xgb, parametreler, cv=10, n_jobs=-1 ,verbose=2)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
# En uygun parametreleri giriyoruz.
xgb1 = XGBRegressor(colsample_bytree = 0.5, learning_rate = 0.09, max_depth = 4,n_estimators=2000)
```

```python
# Modelimizi eğitiyoruz.
model_xgb = xgb1.fit(X_train,y_train)
```

```python
# Tahmin yapıyoruz.
print("Tahmin edilen değerler : \n" ,model_xgb.predict(X_test))
```

```python
print("Gerçek değerler : \n", y_test)
```

```python
# Modelimizin skorunu hesapladık 0-1 arası değerler  döndürür bize.
print("score :", model_xgb.score(X_test,y_test)) # Test seti
```

```python
model_xgb.score(X_train,y_train) # Eğitim seti
```

```python
# Valide edilmiş yani doprulanmış hata oranımızı buluyoruz.
np.sqrt(-1*(cross_val_score(model_xgb, X_test, y_test, cv=10, scoring="neg_mean_squared_error"))).mean()
```

```python
# Modeldeki parametrelerin önemine bakıyoruz.
importance = pd.DataFrame({"Önemli": model_xgb.feature_importances_},
                          index=X_train.columns)
```
