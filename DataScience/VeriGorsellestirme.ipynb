## Veri Setini Tanıma ve Görselleştirme

**Kütüphanelerin Eklenmesi**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Veri Setini Tanıma**

```python
df = pd.read_excel("borsa.xlsx")
pd.DataFrame(df)
```

**Veri Setini Tanıma**

**Firma İncelemesi**

```python
firmalar = df["firma_unvani"].value_counts()
firmalar
```

```python
firmalar.value_counts().plot.barh()
```

**Sektörleri İnceleme**

```python
sektor = df["sektor"].value_counts()
sektor
```

```python
print(df["sektor"].value_counts().plot.barh())
```

**Hissedar Sayılarını İnceleme**

```python
hissedarSayisi = pd.DataFrame(df[["firma_unvani","hissedar_sayisi"]])
hissedarSayisi.value_counts()
df["hissedar_sayisi"].value_counts()
```

```python
fig = plt.figure(figsize=(8,19))
df["hissedar_sayisi"].value_counts().plot(kind = "pie", autopct = "%.f%%");
plt.ylabel("Firmaların Hissedar Oranları", fontsize = 15)
```

**Firmaların Personel Sayıları İncelemesi**

```python
personelSayisi = pd.DataFrame(df[["sektor","personel_sayisi"]])
personelSayisi.value_counts()
```

**Büyüme Oranı İncelemesi**

```python
buyumeOranı = pd.DataFrame(df[["firma_unvani","borsadaki_piyasa_degeri"]])
buyumeOranı.value_counts()
```

**Pazar Payı İncelenmesi**

```python
pazarPayi = pd.DataFrame(df[["firma_unvani","pazar_payi"]])
pazarPayi.value_counts()
```

**Arge'ye Ayırılan Payların incelenmesi**

```python
argePayi = pd.DataFrame(df[["sektor","arge_ye_ayirdigi_pay"]])
argePayi.value_counts()
```

```python
fig = plt.figure(figsize=(20,5))
df["arge_ye_ayirdigi_pay"].value_counts().plot(kind = "pie", autopct = "%.f%%");
plt.ylabel("", fontsize = 15)
plt.title("Arge Payı")
```

**Devlet Desteği İncelenmesi**

```python
df["devlet_destegi"].value_counts()
```

```python
fig = plt.figure(figsize=(20,5))
df["devlet_destegi"].value_counts().plot(kind = "pie", autopct = "%.f%%");
plt.ylabel("", fontsize = 15)
plt.title("Devlet Desteği Oranı")
```

**Gelişmiş reklamcılık faaliyet incelemesi**

```python
argePayi = pd.DataFrame(df[["gelismis_reklamcilik_faaliyetleri"]])
argePayi.value_counts()
```
